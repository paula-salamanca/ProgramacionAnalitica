

<h3>Identificar los valores Nulos</h3>



```python
#Reemplazarlos
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

<h4>Contar los valores nulos</h4>



```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

Las 4 variables van a ser : Tipo Vino, calidad, alcohol, acidez citrica


```python
bins = np.linspace(min(df["quality"]), max(df["quality"]), 4)
bins
```

```python

group_names = ['Low', 'Medium', 'High']
df['quality-binned'] = pd.cut(df['quality'], bins, labels=group_names, include_lowest=True )
df[['quality','quality-binned']].head(20)
```

```python
df["quality-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["quality-binned"])

# set x/y labels and plot title
plt.pyplot.xlabel("Calidad Vino")
plt.pyplot.ylabel("Cantidad vinos")
```

```python
bins = np.linspace(min(df["alcohol"]), max(df["alcohol"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['alcohol-binned'] = pd.cut(df['alcohol'], bins, labels=group_names, include_lowest=True )
df[['alcohol','alcohol-binned']].head(20)
```

```python
df["alcohol-binned"].value_counts()
```
